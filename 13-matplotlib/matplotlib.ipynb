## 1.2 combinate multiple graphs

```python
# set style

plt.style.use("fivethirtyeight")

# generate data for plotting
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# matplotlib.pyplot.subplot(* args, **kwargs)
# add a subplot to the current figure.*args
# args can be threee intergers (nrows, ncols, index),
# or a 3-digit interger. The digits are interpreted as if given sepately as three

plt.subplot(2, 1, 1)  # first subplot
plt.plot(x, y_sin)  # plot the first subplot
plt.title("Sine")  # set tittle for fist subplot
plt.subplot(2, 2, 2)
plt.plot(x, y_cos)
plt.title("cosine")
plt.show()
```

## 1.3 matplotlib custom numerous formatting like color, marker and line style.

```python
plt.style.use("seaborn-v0_8-dark")

# create a 8x6 (inches) figure, dpi means the resolution of the figure (dpi=80 ...)

plt.figure(figsize=(8, 6), dpi=80)

# generate data

X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)

# draw a blue curve with customized format.

plt.plot(X, C, color="blue", linewidth=1.0, label="Blue", linestyle="--")

# draw a grren curve with cusomized format

plt.plot(X, S, color="green", linewidth=1.0, label="Green", linestyle="-.")

plt.legend()  # place a legend on the axes
plt.xlim(-4.0, 4 - 0)  # set a scale of the x-axis
plt.xticks(np.linspace(-4, 4, 9, endpoint=True))  # set the current tick locations ...
plt.ylim(-1.0, 1.0)  # set a scale of the x-axis
plt.yticks(np.linspace(-1, 1, 5, endpoint=True))  # set the current tick locations ...
# plt.savefing('exercice.png', dpi=72) # save figure
plt.show()
```

```python
# CHANGE COLOR AND LINEWIDTH

plt.figure(dpi=80)
plt.plot(X, C, color="blue", linewidth=3.5, linestyle="--")
plt.plot(X, S, color="red", linewidth=3.5, linestyle="-")
plt.show()
```

# 2. scatter plot

```python
## generate data

a = np.random.randint(0, 20, 15)
b = np.random.randint(0, 20, 15)
print(a)
print(b)

# matplotlib.pyplot.scatter(x,y,s=None,c=None,marker=None,cmap=None,norm=None)
# A scatter plot of y vs. x with varyng marker size and/or color

plt.scatter(a, b, c="blue", s=100, marker=".")  # plot scatter
plt.show()
```

# 3. Bar plot 

```python
level = ["Execellent", "god", "soso"]
x = range(len(level))  # x-axis
y = [1, 3, 2]  # y-axis
plt.figure(dpi=100)  # create figure

# matplotlib.pyplot.bar(x, height, width=0.8, bottom=None, *, aling='center')
# make a bar plot

plt.bar(x, y, width=0.5, color=["b", "r", "g"])
plt.xticks(x, level)

# add grid
plt.grid(linestyle="--", alpha=0.5)
plt.show()
```

# 4. Histogram

```python
# generate data
t = np.random.randint(0, 30, 90)
print(t)

# create figure
plt.figure(dpi=100)

# set group number(i-e. the range of each bin)

group_num = 14

# matplotlib.pyplot.hist(x,bins=None, range = None, destiny=False, wights=None)
# plot a histogram. x indicates the data. bins indicates the number of bins

plt.hist(t, facecolor="blue", edgecolor="black", bins=group_num, alpha=1)
plt.xticks(range(min(t), max(t))[::2])
plt.grid(linestyle="--", alpha=0.5)
plt.show()
```
