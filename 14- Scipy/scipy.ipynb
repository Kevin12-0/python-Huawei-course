## Discrete fourier transforms (scipy.fft)

# 2.1 Fats fourier Transforms (FFTs)

FFT (Fast Fourier Transform) referes ti a way the discrete Fourier Transform (DFT) can be calculated efficiently, by usig symetires in the calculated terms. The symmetry is highest when n is a power of 2, and the transform if therefore most efficent for sizes. For poorly factorizalble sizes, scipy.fft uses BluesteinÂ´s algorithm and so is never worser than 0(n long n).

*   fft(x[,n.axis,norm,overwrite_x,...]):Compute the 1-D discrete Fourier Transform.
*   ifft(x[,n,axis,norm,overwrite_xx,...]):Compute the 1-D inverse discrete Fourier Transform.

```python
import scipy.fft
import numpy as np

a = np.exp(2j * np.pi * np.arange(8) / 8)
print(a)
b = scipy.fft.fft(a)
print(b)
```

```python
scipy.fft.ifft(b)
```

# 3. interpolation (scipy.interpolate)

sub-package for objects used in interpolation

As listened below, this sub-packegae contains spline funtions and clases, 1-D and multidemcional (univariate and multivariate) interpolation classe, Lagrange and Tylor polynomial interpolators, and wrappers for FITPACK and DFITPACK funtions.

*   interp1d(x,y[, kind,axis,copy,...]): interpolate a 1-D funtion
*   UnivariateSpline(x,y[,w,bbox,k,s,ext,...]): 1-D smoothing fit to a given set of data points.

```python
from scipy import interpolate as intp
import matplotlib.pyplot as plt

x = np.linspace(0, 4, 12)
y = np.cos(x**2 / 3 + 4)
plt.plot(x, y, "o")
plt.show()
```

# 31. scipy.interpolate.inerp1d

scipy.interpolate.interp1d(x,y,kind='linear', axis=-1,copy=True,bound_error_None,fill_value=nan,assume_sorted=False)

Interpolate a 1-D funtion.

*   x and y arrays of values used to approximate some funtions f:y = f(x). This class retunrs a funtion whose call method used interpolation to find the value of new points.
*   kind: Specufues the kind of interpolation as a string ('linear','nearest','zero','slinear','quadratic','cubic','previous','next','where zero','slinear','quadratic' and 'cubic' refer to a spline interpolation of a zeroth, fisrst, second or third ordes; 'previous' and 'next' symply return the precious or next value of the point) or as interger specifyng the order of the spline interpolation to use. Default is 'linear'.


```python
f1 = intp.interp1d(x, y, kind="linear")
f2 = intp.interp1d(x, y, kind="cubic")
xnew = np.linspace(0, 4, 300)
plt.plot(x, y, "o", xnew, f1(xnew), "-", xnew, f2(xnew), "--")
plt.legend(["data", "linear", "cubic", "nearest"], loc="best")
plt.show()
```

# 3.2 scipy.interpolate.UnivariateSpline

scipy.interpolate.UnivariateSplite(x,y,w=None,bbox=[None,None],k=3,s=None,ext=0,check_finite=False)

1-D smoothing spline fit to a given set of data points.

Methods:

*   set_smoothing_facor(self,s): Continue spline computation eith the given smoothing factor s and with knots found at the last call. This routine modifies the spline in place.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import UnivariateSpline

x = np.linspace(-3, 3, 50)
y = np.exp(-(x**2)) + 0.1 * np.random.randn(50)  ## add noise by random method
plt.plot(x, y, "ro", ms=5)
```

```python
plt.plot(x, y, "ro", ms=5)

xs = np.linspace(-3, 3, 1000)
spl = UnivariateSpline(x, y)
# set_smoothing_factor as 0
spl.set_smoothing_factor(0)
plt.plot(xs, spl(xs), "yellow", lw=3)  # yellow line
```

```python
plt.plot(x, y, "ro", ms=5)
# set_smoothing-factor as 0.5
spl.set_smoothing_factor(0.5)
plt.plot(xs, spl(xs), "green", lw=3)  # green line

plt.show()
```

```python
plt.plot(x, y, "ro", ms=5)
# set_smoothing_factor as 100
spl.set_smoothing_factor(100)
plt.plot(xs, spl(xs), "blue", lw=3)  # blue line
```

# 5.2 Singular value decomposition

singular value descomposition (svd) can be thought of as an extension of the eigenvalue problem to matrices that are not square. Let A and M x N Matrix with M and N arbitrary. The matrices A^H A and AA^H are square hermitian matrices of size N x N and M x M, respectively. It is Known that the eigenvalues of square hermitian matrices are real and non negative. In addition, there are at most min(M,N) identical non-zero eigenvalues of A^H and A and AA^H. Define these positive eigen values as o^2. The square root of these are called singular values of A. The igenvectors of A^H A are collected by columns into NxN Unirary matrix V, while the eigenvectors of AA^H are collected by columns in the unitary matrix U, the singular values are collected in an MxN zero matrix E with main diagonal set to the singular vales. Then is the singular values decomposition of A. Every matrix has singular value descomposition. Sometimes the singular values are called the spectrum of A. The command lingalg.svd will retunr U,V^H, and o, as an array of singular values. To obtain the matrix E, use lingalg.diagsvd. The following example illustrates the use of lingalg.svd:

```python
A = np.array([[1, 2, 3], [4, 5, 6]])
A
```

```python
M, N = A.shape
U, s, Vh = linalg.svd(A)
Sig = linalg.diagsvd(s, M, N)
U
```
